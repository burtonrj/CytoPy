```python
import matplotlib.pyplot as plt
target_sample_pca = pd.read_csv('target.csv').values
projection_before = pd.read_csv('before.csv').values
projection_after = pd.read_csv('after.csv').values
fig, ax = plt.subplots()
ax.scatter(target_sample_pca[:,0], target_sample_pca[:,1], c='blue', s=3, alpha=0.1)
ax.scatter(projection_before[:,0], projection_before[:,1], c='red', s=3, alpha=0.1)
ax.scatter(projection_after[:,0], projection_after[:,1], c='green', s=3, alpha=0.1)
plt.show()
```
