```python
model = DBSCAN(eps=0.08, min_samples=10).fit(training_data[features])
```

```python
set(model.labels_)
```

```python
training_data['labels'] = model.labels_
```

```python
fitter = UMAP()
```

```python
embeddings = fitter.fit_transform(training_data[features])
```

```python
training_data['umap0'] = embeddings[:, 0]
training_data['umap1'] = embeddings[:, 1]
```

```python
fig, ax = plt.subplots(figsize=(10,10))
with sns.color_palette("muted"):
    sns.scatterplot(data=training_data, x='umap0', y='umap1', hue='labels', 
                    s=8, alpha=0.5, ax=ax)
```
